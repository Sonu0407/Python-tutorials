```python
def factorial_recursive(n):
    """Calculate factorial using recursion"""
    if n == 0:
        result = 1
    else:
        result = n * factorial_recursive(n - 1)# 5 * 4 
    return result

n = 5
res = factorial_recursive(n)
print(f'The factorial of {n} is {res}')  # 120
```

```python
class car:
    def __init__(self, name, model, color):
        self.name = name
        self.model = model
        self.color = color

    def display(self):
        print(f"Name: {self.name}")
        print(f"Model: {self.model}")
        print(f"Color: {self.color}")

# creating the object 
# object_name = maruthi_800
# class_name = car

maruthi_800 = car(name='maruthi_800',model='1 stroke',color='white')

# accessing the attributes and method

# object_name.attribute
print(maruthi_800.color)
maruthi_800.display()
print('*'*50)
```

```python
# example for method overriding 

class parent:

    def car(self):
        print("maruti 800")

    def land(self):
        print("doing agriculture")

class child(parent):

    def moblie(self):
        print("Iphone")

    def land(self):
        print("appartment")

    def car(self):
        print("modified maruthi 800")

c1 = child()
c1.moblie()
c1.land()
c1.car() 
 
```
