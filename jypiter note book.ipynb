```python
def print_kwargs(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_kwargs(name="sonu", course="Fullstack Python")
print_kwargs(name="sonu")
print_kwargs(name="Amaresh", course="Fullstack Java", starter="Front-End")
```

```python
def even_generator(limit):
    for i in range(2, limit+1, 2):
        yield i

for num in even_generator(10):
    print(num)
```

```python
class Car:
    def __init__(self, Brand, model):
        self.Brand = Brand
        self.model = model

    def display(self):
        print(f"Brand = {self.Brand}")
        print(f"model = {self.model}")

my_car = Car("toyota","1988")
my_car.display()
```

```python
# Inheritants

class Car:
    total_cars = 0
    def __init__(self, Brand, model):
        self.__Brand = Brand # Encapsulation __private _protected normal = public
        self.__model = model
        Car.total_cars += 1

    def display_full(self):
        return (f"Name of car = {self.__Brand} and Model is {self.__model}")

    def fuel_type(self): #polymorphism
        return "Petrol or Diesel"
    
    @staticmethod # it does not rely on object and instance to run this function 
    def general_discription(): #static method
        return "Car means for transport"
    
    @property  # from you can't change the car model name it only read-only
    def model(self):
        return self.__model
    
class ElectricCar(Car): # inheritance
    def __init__(self, Brand, model, Battery_size): 
        super().__init__(Brand, model)
        self.Battery_size = Battery_size

    def fuel_type(self):
        return "Electric Charge"
    

my_tesla = ElectricCar("Tesla","Model S","80kwh")
print(my_tesla.display_full())
my_car = Car("BMW","Rolls Royce")
print(my_tesla.fuel_type())
print(my_car.fuel_type())
print(Car.total_cars)
print(my_car.general_discription())
print(Car.general_discription())
# my_car.model="Ghost" gave error # you can modify model untill you used @property method in Car class
print(my_car.display_full())
print(isinstance(my_tesla, Car)) # checks whether it is an instance of Car class Prove it!
print(isinstance(my_tesla, ElectricCar)) # checks whether it is also instance of Electric Class Prove it!
```
